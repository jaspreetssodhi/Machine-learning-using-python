### This graph represent the majority income of the people & majority of them have income between 61k to 80k dollars

```python
#THIS REPRESENTS THE SPENDING SCORE OF THE CUSTOMERS
#MOST OF THE PEOPLE HAVE THIER SPENDING SCORE BETWEEN 41 TO 50 AND 51 TO 60

s0_10 = mall["Spending Score (1-100)"][ (mall["Spending Score (1-100)"]>0) & ( mall["Spending Score (1-100)"]<=10) ]
s11_20 = mall["Spending Score (1-100)"][ (mall["Spending Score (1-100)"]>10) & ( mall["Spending Score (1-100)"]<=20) ]
s21_30 = mall["Spending Score (1-100)"][ (mall["Spending Score (1-100)"]>20) & ( mall["Spending Score (1-100)"]<=30) ]
s31_40 = mall["Spending Score (1-100)"][ (mall["Spending Score (1-100)"]>30) & ( mall["Spending Score (1-100)"]<=40) ]
s41_50 = mall["Spending Score (1-100)"][ (mall["Spending Score (1-100)"]>40) & ( mall["Spending Score (1-100)"]<=50) ]
s51_60 = mall["Spending Score (1-100)"][ (mall["Spending Score (1-100)"]>50) & ( mall["Spending Score (1-100)"]<=60) ]
s61_70 = mall["Spending Score (1-100)"][ (mall["Spending Score (1-100)"]>60) & ( mall["Spending Score (1-100)"]<=70) ]
s71_80 = mall["Spending Score (1-100)"][ (mall["Spending Score (1-100)"]>70) & ( mall["Spending Score (1-100)"]<=80) ]
s81_90 = mall["Spending Score (1-100)"][ (mall["Spending Score (1-100)"]>80) & ( mall["Spending Score (1-100)"]<=90) ]
s91_100 = mall["Spending Score (1-100)"][ (mall["Spending Score (1-100)"]>90) & ( mall["Spending Score (1-100)"]<=100) ]

x=["0 to 10","11 to 20","21 to 30","31 to 40","41 to 50","51 to 60","61 to 70","71 to 80","81 to 90","91 to 100"]
y=[len(s0_10.values),len(s11_20.values),len(s21_30.values),len(s31_40.values),len(s41_50.values),len(s51_60.values),len(s61_70.values),len(s71_80.values),len(s81_90.values),len(s91_100.values)]

plt.figure(figsize=(10,6))
plt.grid()
sns.barplot(x=x,y=y, palette = "Set1")
plt.title("Spending score vs Count")
plt.xlabel("Spending Score (0 to 100)")
plt.ylabel("Number of people")
plt.show()
```

### The above represnts the spending score of the customers,  majority of them have their spending score between 41 to 60


```python
#WCSS means Within-Cluster-Sum-of-Squares 
#we use WCSS to find out the most appropriate number of cluster
#At the point after which there is no significant drop in the graph is known as elbow point and it is selcted as the best number of clusters

wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(mall.iloc[:,1:])
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss, linewidth=2, color="red", marker ="8")
plt.xlabel("Number pf cluster")
plt.ylabel("WCSS")
```

## The above graph represents the most appropriate numbers of clusters i.e. 5 cluster

```python
km = KMeans(n_clusters=5)
clusters = km.fit_predict(mall.iloc[:,1:])
mall["label"] = clusters

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
 
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(mall.Age[mall.label == 0], mall["Annual Income (k$)"][mall.label == 0], mall["Spending Score (1-100)"][mall.label == 0], c='blue', s=60)
ax.scatter(mall.Age[mall.label == 1], mall["Annual Income (k$)"][mall.label == 1], mall["Spending Score (1-100)"][mall.label == 1], c='red', s=60)
ax.scatter(mall.Age[mall.label == 2], mall["Annual Income (k$)"][mall.label == 2], mall["Spending Score (1-100)"][mall.label == 2], c='green', s=60)
ax.scatter(mall.Age[mall.label == 3], mall["Annual Income (k$)"][mall.label == 3], mall["Spending Score (1-100)"][mall.label == 3], c='orange', s=60)
ax.scatter(mall.Age[mall.label == 4], mall["Annual Income (k$)"][mall.label == 4], mall["Spending Score (1-100)"][mall.label == 4], c='purple', s=60)
#ax.view_init(30, 185)
plt.xlabel("Age")
plt.ylabel("Annual Income (k$)")
ax.set_zlabel('Spending Score (1-100)')
plt.show()
```
